```python
# PEDIR UM JWT
from src.rest_api.rest_build import RestBuilder

DOMAIN = "http://localhost"

headers_dict = {
    'Content-Type' : 'application/json'
}
body_dict = {
    'username' : 'rafael.freitas@recife.pre.gov.br', 
    'password' : '123456'
}

b = RestBuilder( domain=DOMAIN )
# b.m_debug_on()

b.m_set_method(b.METHOD_POST)
b.m_set_endpoint("api/v1/auth")
b.m_set_port("8080")
b.m_set_headers(headers_dict=headers_dict)
b.m_set_body(body_dict=body_dict)
b.m_to_string()
saida = b.m_send_request()
token = saida['token']
token

```
